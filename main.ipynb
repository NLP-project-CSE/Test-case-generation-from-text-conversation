```python
def get_keywords(text):
    out=[]
    try:
        # extractor = pke.unsupervised.MultipartiteRank()
        extractor = pke.unsupervised.YAKE()
        pos = {'VERB', 'ADJ', 'NOUN'}
        #pos ={'NOUN'}
        stoplist = list(string.punctuation)
        stoplist += ['-lrb-', '-rrb-', '-lcb-', '-rcb-', '-lsb-', '-rsb-']
        stoplist += stopwords.words('english')
        extractor.load_document(input=text,
                                stoplist=stoplist)
        extractor.candidate_selection(n=1)

        extractor.candidate_weighting(window=3
                                      ,use_stems=False)

        keyphrases = extractor.get_n_best(n=30)
        

        for val in keyphrases:
            out.append(val[0])
    except:
        out = []
        traceback.print_exc()

    return out

keywords = get_keywords(text)[:8]
print ("keywords: ",keywords)
```

```python
sentences = tokenize_sentences(text)
print (sentences)
```

```python
from pprint import pprint
def get_sentences_for_keyword(keywords, sentences):
    keyword_processor = KeywordProcessor()
    keyword_sentences = {}
    for word in keywords:
        keyword_sentences[word] = []
        keyword_processor.add_keyword(word)
    for sentence in sentences:
        keywords_found = keyword_processor.extract_keywords(sentence)
        for key in keywords_found:
            keyword_sentences[key].append(sentence)

    for key in keyword_sentences.keys():
        values = keyword_sentences[key]
        values = sorted(values, key=len, reverse=False)
        keyword_sentences[key] = values
    return keyword_sentences

keyword_sentence_mapping = get_sentences_for_keyword(keywords, sentences)
pprint (keyword_sentence_mapping)
# remove all the reapeated sentences in keyword_sentence_mapping
for key in keyword_sentence_mapping.keys():
    values = keyword_sentence_mapping[key]
    values = list(set(values))
    keyword_sentence_mapping[key] = values
pprint (keyword_sentence_mapping)

#save keywords_sentence_mapping in csv file
import csv
with open('keywords_sentence_mapping.csv', 'w') as csvfile:
    fieldnames = ['keyword', 'sentence']
    writer = csv.DictWriter(csvfile, fieldnames=fieldnames)
    writer.writeheader()
    for key in keyword_sentence_mapping.keys():
        for sentence in keyword_sentence_mapping[key]:
            writer.writerow({'keyword': key, 'sentence': sentence})
```

```python
# get 20 random sentences for each keyword
def get_random_sentences(keyword_sentence_mapping):
    random_sentences = {}
    for key in keyword_sentence_mapping.keys():
        random_sentences[key] = keyword_sentence_mapping[key][:5]
    return random_sentences
get_random_sentences(keyword_sentence_mapping)

```
